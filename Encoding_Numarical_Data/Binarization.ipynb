```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import Binarizer
from sklearn.compose  import ColumnTransformer
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("Train.csv",usecols=['Age','Fare','SibSp','Parch','Survived'])
df.head()
```

```python
df['family'] = df['SibSp'] + df['Parch']
```

```python
df.drop(columns=['SibSp','Parch'],inplace=True)
df.dropna(inplace=True)
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=30)
```

```python
# without binarization
cl=DecisionTreeClassifier()
cl.fit(X_train,y_train)
y_pred=cl.predict(X_test)
acc=accuracy_score(y_test,y_pred)
acc
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),X,y,cv=8,scoring='accuracy'))
```

```python
#with bbinarization
trf=ColumnTransformer([('binary',Binarizer(copy=True),['family'])],remainder='passthrough')

trf = ColumnTransformer([
    ('bin',Binarizer(copy=False),['family'])
],remainder='passthrough')
```

```python
X_train_trf=trf.fit_transform(X_train)
X_test_trf=trf.fit_transform(X_test)

pd.DataFrame(X_train_trf,columns=['family','Age','Fare'])
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_trf,y_train)
y_pred2 = clf.predict(X_test_trf)

accuracy_score(y_test,y_pred2)
```

```python
X_trf = trf.fit_transform(X)
np.mean(cross_val_score(DecisionTreeClassifier(),X_trf,y,cv=8,scoring='accuracy'))
```
